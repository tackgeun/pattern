### LOAD Iris dataset

```python
%matplotlib inline
import numpy as np
from pylab import *
import matplotlib.pyplot as plt
```

```python
# Load data
def load_data(filename):
    file = open(filename, "rb")
    lines = file.readlines()
    features = [(map(float, x.split(',')[:-1])) for x in lines[:-1]]
    labels = [x.split(',')[-1].strip() for x in lines[:-1]]
    
    return features, labels
```

```python
[feats, labels] = load_data('iris.data')
#feats=np.array(feats)
#labels=np.array(labels)

x1=[];x2=[];x3=[]
y1=[];y2=[];y3=[]

for feat, label in zip(feats, labels):
    if label == 'Iris-setosa':
        x1.append(feat)
        y1.append(label)
    elif label == 'Iris-versicolor':
        x2.append(feat)
        y2.append(label)
    elif label == 'Iris-virginica':
        x3.append(feat)
        y3.append(label)
        
x1=np.array(x1)
x2=np.array(x2)
x3=np.array(x3)
y1=np.array(y1)
y2=np.array(y2)
y3=np.array(y3)
```

```python
# subplots grid
groups = unique(labels)
colors = ("blue", "green", "red")


fig = plt.figure(figsize=(10, 8))

for i in range(0,4):
  for j in range(0,4):
    idx = j*4 + i+1
    if i==j:
      data = [x1[:,i], x2[:,i], x3[:,i]]
      for dat, color, group in zip(data, colors, groups):
        plt.subplot(4,4,idx)
        plt.hist(dat,10, edgecolor='black')
    else:
      data = [(x1[:,i],x1[:,j]), (x2[:,i],x2[:,j]), (x3[:,i],x3[:,j])]
      for dat, color, group in zip(data, colors, groups):
        x,y=dat
        plt.subplot(4,4,idx)
        plt.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

plt.show()
```
